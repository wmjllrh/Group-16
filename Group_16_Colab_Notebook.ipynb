###Cleaning and joining datasets

####Election data (2020)

```python
# Removing all years bar 2020
election_data.drop(election_data[election_data['year'] != 2020].index, inplace = True)

# Removing unnecessary columns
election_data = election_data.drop (columns = [election_data.columns[2], election_data.columns[4], election_data.columns[5], 
                                               election_data.columns[6], election_data.columns[10], election_data.columns[11]])

# Converting 'party' row to column headers
ed = election_data.pivot_table('candidatevotes', ['year', 'state', 'county_name', 'totalvotes'], 'party', aggfunc = 'sum')

# Collapsing rows (so that each row represents a state)
ed = ed.groupby('state').sum()

# Re-creating a 'total votes' column
ed['total_votes'] = ed['DEMOCRAT'] + ed['GREEN'] + ed['LIBERTARIAN'] + ed['OTHER'] + ed['REPUBLICAN']

# Creating percentage vote columns (so results are standardised and comparable)
ed['democrat_pct_2020'] = ed.DEMOCRAT / ed.total_votes * 100
ed['green_pct_2020'] = ed.GREEN / ed.total_votes * 100
ed['libertarian_pct_2020'] = ed.LIBERTARIAN / ed.total_votes * 100
ed['other_pct_2020'] = ed.OTHER / ed.total_votes * 100
ed['republican_pct_2020'] = ed.REPUBLICAN / ed.total_votes * 100

# Retaining only the newly created columns
ed = ed.iloc[:, 6:11]

# Re-ordering columns
ed = ed[['republican_pct_2020', 'democrat_pct_2020','green_pct_2020','libertarian_pct_2020','other_pct_2020']]
```

####Gun Violence Archive data (2021)

```python
# Joining GVA data
gva_data = pd.concat ([jan,feb,mar,apr,may,jun,jul,aug,sep,oct,nov,dec], ignore_index = True)

# Grouping GVA data by state, then summing the number of mortalities by firearms in each.
gva_data = gva_data.groupby ('State') ['# Killed'].sum().reset_index()

# Renaming the '# Killed' column and 'State' column
gva_data.rename (columns = {'# Killed':'deaths_2021', 'State':'state'}, inplace = True)

# Setting 'state' as the index
gva_data = gva_data.set_index('state')
```

####US Census population estimates data (2021)

```python
# Renaming US Census population estimates 'state' and the 2021 population estimate column
pop.rename (columns = {pop.columns[0]: "state", pop.columns[3]: "population_2021"}, inplace = True)

# Removing 2020 population estimates columns. For our analyses, we will use 2021 population estimates.
pop = pop.drop (columns = [pop.columns[1], pop.columns[2]])

# Setting 'state' as the index
pop = pop.set_index('state')
```

####US decennial ethnicity census data [2020: DEC Redistricting Data (PL94-171)]

```python
# Renaming columns
eth.rename (columns =
            {eth.columns[0]: "state", eth.columns[1]: "total", eth.columns[3]:"white_alone_pop"},
            inplace = True)

# Shift state column rows down by ONE
eth.loc[:, 'state'] = eth.state.shift(+1)

# Removing ethnicity data rows with NaN values and removing commas
eth = eth.dropna()
eth = eth.replace(',','', regex = True)

# Retaining only state, white pop and total columns
eth = eth[['state', 'white_alone_pop', 'total']]

# Setting 'state' as the index
eth = eth.set_index('state')

# Converting remaining columns to numeric
eth = eth.apply(pd.to_numeric)

# Creating a % 'white alone' column
eth['white_alone_pct'] = eth.white_alone_pop / eth.total * 100

# Dropping unnecessary columns
eth = eth.drop (columns = [eth.columns[0], eth.columns[1]])
```

####American Community Survey 2020 5-Year Estimates - Income in the past 12 months (IN 2020 INFLATION-ADJUSTED DOLLARS)

```python
# Renaming columns of interest
income_data.rename (columns =
            {income_data.columns[0]: "state", income_data.columns[2]: "household_less_than_10k", income_data.columns[11]:
             "household_200k_or_more", income_data.columns[12]: "median_household_income"},
            inplace = True)

# Shift state column rows down by TWO
income_data.loc[:, 'state'] = income_data.state.shift(+2)

# Retaining only columns of interest
income_data = income_data[['state', 'household_less_than_10k', 'household_200k_or_more', 'median_household_income']]

# Removing data rows with NaN values and removing commas
income_data = income_data.dropna()
income_data = income_data.replace(',','', regex = True)

# Converting percentage strings to numeric
income_data['household_less_than_10k'] = income_data['household_less_than_10k'].str.rstrip("%").astype(float)
income_data['household_200k_or_more'] = income_data['household_200k_or_more'].str.rstrip("%").astype(float)

# Setting 'state' as the index
income_data = income_data.set_index('state')

# Converting remaining columns to numeric
income_data = income_data.apply(pd.to_numeric)

# Making the index uppercase to match our combine dataframe
income_data.index = income_data.index.str.upper()
```

####State abbreviations

```python
# Renaming columns
abbreviations.rename (columns =
            {abbreviations.columns[0]: "state", abbreviations.columns[1]: "code"},
            inplace = True)

# Setting 'state' as the index
abbreviations = abbreviations.set_index('state')
```

####Combined data

```python
# Left joining the GVA and Census datasets (keeping only the index values of 'gva_data')
df = gva_data.merge (pop, left_index = True, right_index = True, how = 'left')

# Removing commas from the dataframe and converting the 'population_2021' column to numeric
df = df.replace(',','', regex = True)
df["population_2021"] = pd.to_numeric(df["population_2021"])

# Standardising firearms mortalities to allow comparison across states 
## [creating a 'rate of firearms-related mortalties per 100,000 people' column]
df['deaths_per_100k'] = df.deaths_2021 / df.population_2021 * 100000

# Removing unnecessary 'population' and 'mortality' columns
df = df.drop (columns = [df.columns[0], df.columns[1]])

# Merging ethnicity data
df = df.merge (eth, left_index = True, right_index = True, how = 'left')

# Making the index uppercase, so we may join the election data
df.index = df.index.str.upper()

# Merging 2020 election data
df = df.merge (ed, left_index = True, right_index = True, how = 'left')

# Merging 2020 ACS income estimates data
df = df.merge (income_data, left_index = True, right_index = True, how = 'left')

# Merging state abbreviations
df = df.merge (abbreviations, left_index = True, right_index = True, how = 'left')

# Creating a categorical variable to indicate whether the vote share percentage of a state was higher for democrats or republicans
conditions = [
    (df['republican_pct_2020'] > df['democrat_pct_2020']),
    (df['republican_pct_2020'] < df['democrat_pct_2020']),
]
values = ['Republican', 'Democrat']
df['vote_share_maj'] = np.select(conditions, values)

# Removing the district of columbia
df_states = df.drop ('DISTRICT OF COLUMBIA')
```

##Checking regression assumptions
When evaluating the presented model used for multiple regression, it is important to consider various regression 'assumptions', as these allow for a statistical indication of both the reliability and validity of the results produced. As such, if the relationship between independent variables () and the dependent variable () is not linear, and errors do not follow a normal distribution, we might expect our model estimates to be **inefficient**, wherein we see much greater variances and reduced precision compared to what we would see if such assumptions were met. Further, if residuals are **not** independent and homoskedastic (*ie, there are systematic correlations and non-constant variances observed*), then our model may be **systematically biased** - ultimately, creating results that are under or over estimated (Edwards, 1977). As such, in this section of the website, each formal regression assumption is examined both substantively and statistically for the presented model used in multiple regression, then an updated model (*including various statistical techniques that may better help us satisfy these assumptions where appropriate*) is presented in a geographically weighted regression analysis utilsing R. 

Whilst multicollinearity is not necessarily a formal assumption of regression, both inefficiencies and biases may be presented if this factor remains unchecked when considering the presented model. This is because multicollinearity makes it difficult to distinguish the influence of each variable independently (*in relation to the dependent variable*). As such, potential multicollinearity is also examined in this section.

###Is the relationship between each x and y linear?

For linear regression to be useful, we must first check whether the relationship between each independent and dependent variable is truly linear. We will assess this by utilising scatterplots, and calculating Pearsons correlation coefficients to indicate the level of linearity.

####% of Republican Vote Share

```python
# Returning pearsons correlation coefficient
corr, _ = pearsonr(df_states["republican_pct_2020"], df_states["deaths_per_100k"])
correlation = 'Pearsons Correlation Coefficient: %.3f' % corr

# Plotting the variables
colours = np.where(df_states["vote_share_maj"].values == "Democrat", "#0000FF", "#FF0000")
fig, ax = plt.subplots(figsize=(13.5, 10))
scatter = ax.scatter(df_states["republican_pct_2020"].values,
                     df_states["deaths_per_100k"].values,
                     c = colours,
                     s = 0)

# Add text labels for each scatter point
text_labels = []  # list to store the Text objects
for i, txt in enumerate(df_states.index):
    jitter_x = random.uniform(-0.2, 0.2)
    jitter_y = random.uniform(-0.2, 0.2)
    text = ax.text(df_states["republican_pct_2020"].values[i] + jitter_x,
                   df_states["deaths_per_100k"].values[i] + jitter_y,
                   txt,
                   horizontalalignment='center',
                   verticalalignment='center',
                   color = colours[i])  
    text_labels.append(text)

# Adjust the position of each text label to avoid overlap
for i, txt in enumerate(text_labels):
    x, y = txt.get_position() 
    for j, other_txt in enumerate(text_labels):
        if i == j:
            continue
        other_x, other_y = other_txt.get_position()
        if abs(x - other_x) < 0.2 and abs(y - other_y) < 0.2:
            txt.set_position((x + random.uniform(-0.9, 0.9), y + random.uniform(-0.9, 0.9)))

plt.title("Gun Violence Mortality Rate [2021] vs Republican Vote Share [2020]", fontsize = 17.5, fontweight = 'bold')
plt.xlabel("Republican Vote Share (2020)", fontsize = 15, fontweight = 'bold')
plt.ylabel("Gun Violence Mortality Rate (2021)", fontsize = 15, fontweight = 'bold')
ax.grid(False)
fig.text(.6, .04, "Produced using data from the Gun Violence Association (2021) and MIT Election Data and Science Lab (2022)", ha='center', style = 'italic')
fig.text(0.79, 0.86, correlation, ha='center', style = 'italic', fontsize = 10)

# Show and save plot
plt.savefig("RepVote.png", dpi = 300, bbox_inches = "tight")
plt.show()
```

It appears that there is a weak positive linear relationship between Republican Vote Share and Gun Violence Mortality Rate. However, a simple linear line (*as explored in the liniear regression section of the website*) may not be the best approach to describe this relationship.  Moreover, it  appears that the relationship might be *curviliniear*. To account for this, we might explore polynomial regression. Thus, we will add a squared term for 'Republican Vote Share' in an attempt to better capture the potential curviliniear relationship ($+\beta x^2$).

###Independence of Observations


One of the first assumptions of regression is that residuals and independent variables are truly independent. 

As all independent variables (*the proportion of the states population identifing as 'White alone' race, the proportion of the states valid votes for the Republican party candidate Donald Trump, and the proportion of households in the state with an annual income of less than $10,000 dollars per year*) were measured in 2020, and the outcome variable (*the rate of gun violence deaths [shot dead] in the state per 100,000 people*) was measured in 2021, we might be confident that the independent variables should not be influenced by the value of the dependent variable, as each independent variable was measured prior to the outcome variable (y) occurring. Also, further confidence that observations are independent may be derived from the fact that each observation is considered on a state level, and thus each observation is independent through geographical distinction.

However, we might test the independence of residuals specifically through use of the Durbin-Watson statistical test (∑(e[i] - e[i-1])^2 / ∑e[i]^2). This permits assessment of the independence assumption regarding the presented model. The test is based on the idea that if the  residuals in the model are independent of one another, autocorrelation between residuals should be close to zero.

```python
# Durbin-Watson Test
dur_wat_test = sms.stats.stattools.durbin_watson(model.resid)

print(dur_wat_test)
```

As such, we do see a small amount of positive autocorrelation in the data (according to the statistic of 2.29). This means that residuals that follow each other in the series tend to be slightly similar in magnitude.

Considering a null hypothesis (*H0 = no autocorrelation*) and the upper and lower boundaries of the Durbin-Watson t-test (*n = 51, k = 3, dL = 1.245, dU = 1.491*), we fail to reject the null hypothesis that there is no autocorrelation presented by the variables within our model. Thus, we might be confident that the indepence assumption is satisfied for the purposes of evaulating our model.
